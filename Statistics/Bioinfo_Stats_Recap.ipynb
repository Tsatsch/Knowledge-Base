```python
import pandas as panda
import matplotlib.pyplot as plot
import numpy as np

from scipy import stats
```

## Mean, Median and Mode

```python
arr = np.array([1, 5, 37, 6, 8, 30, 2, 12, 212, 3, 51, 45, 28, 43, 57, 34, 120, 98, 20, 13, 13])
# Mean
mean = np.mean(arr)
print(mean) 

# Median
median = np.median(arr)
print(median)

# Mode
mode = stats.mode(arr)[0][0]
print(mode)
```

## Variance, SD and SEM

```python
arr = np.array([1, 5, 37, 6, 8, 30, 2, 12, 212, 3, 51, 45, 28, 43, 57, 34, 120, 98, 20, 13, 13])

# Variance
variance = np.var(arr)
print(variance)

# Standard Deviation
std = np.std(arr)
print(std)

# Standard error of the Mean
sem = stats.sem(arr)
print(sem)
```

## Covariance and Correlation

```python
x = np.array([1, 5, 37, 6, 8, 30, 2, 12, 212, 3, 51, 45, 28, 43, 57, 34, 120, 98, 20, 13, 13])
y = np.array([2, 7, 40, 10, 16, 45, 5, 17, 260, 4, 56, 50, 33, 51, 69, 42, 134, 102, 25, 18, 18])

# here should be positive variance: x bigger => y bigger
cov = np.cov(x,y)[0][1] 
print(cov)

# Pearson Correlation: cov(x,y) / std(x) * std(y)
corr = int(np.cov(x,y)[0][1]) / (np.std(x) * np.std(y))
print(corr)
# or alternative
print(stats.pearsonr(x,y)[0])

# Spearman Correlation:
print(stats.spearmanr(x,y)[0])
```

## Normal Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

## QQ Plots


 QQ plots help us to say whether a distribution is normal or not (or maybe exponential or not)

```python
# Example of normal distributed data
import numpy as np
import statsmodels.api as sm
from matplotlib import pyplot as plt

random_normals = [np.random.normal() for i in range(1000)] # array with 1000 random numbers
sm.qqplot(np.array(random_normals), line='45') # create qq plot and draw line 45 degrees
plt.show()
```

See how our data (the blue dots) fall pretty cleanly on the red line? That means that our data is normally distributed. If our data adheres to the red 45 degree line, it’s normal or close to it, and if it does not, then it’s not normal.

```python
# Example of non-normal distributed data
import random

random_normals = [random.random() for i in range(1000)]
sm.qqplot(np.array(random_normals), line='45')
plt.show()
```

Our data (the blue dots) is nowhere close to the red line, meaning it’s not normally distributed (it’s uniformly distributed)
